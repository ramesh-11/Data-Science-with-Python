# Pivoting
The pivot() method helps us restructure a dataset.
it reshapes data based on column values. This method uses unique values from given columns to form axes of the new DataFrame.

```python
import numpy as np
import pandas as pd

from pandas import Series,DataFrame
```

```python
data = pd.read_csv('data_pivot.csv')
data
```

```python
df = data.drop('name',axis=1)
df
```

```python
df.pivot('color','day','value')
```
